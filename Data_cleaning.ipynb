```python
import pandas as pd
```

```python
# List selected reports

reports = [ './/Resources/JC-202401-citibike-tripdata.csv',
            './/Resources/JC-202402-citibike-tripdata.csv',
            './/Resources/JC-202403-citibike-tripdata.csv',]
```

```python
# Empty dataframe

citibike_df = pd.DataFrame()
```

```python
# Loop through & concatonate files

for file in reports:
    data = pd.read_csv(file)
    citibike_df = pd.concat([citibike_df, data], axis=0)

citibike_df
```

```python
citibike_df.to_csv('.///Clean_data.csv', index=False)
```
